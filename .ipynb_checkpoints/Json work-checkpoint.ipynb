```python
import csv
import sys
from collections import defaultdict
import json
import time
```

```python
infile = open('distance.csv')
reader = csv.reader(infile)
d_list = defaultdict(list)
dict = {}
row1 = next(reader)
```

```python
for x in reader:
    distance = float(x[2])
    d=x[0]
    p='%Y-%m-%d'
    epoch = int(time.mktime(time.strptime(d,p)) * 1000) 
    tuplist = [epoch,distance]
    name = x[1]
    d_list[name].append(tuplist)
    print tuplist


   
    
```

```python
print d_list
```

```python
json.dumps(d_list)
print json.dumps(d_list,indent=4, separators=(',', ': '))
```

```python
outfile = open('distance.json', 'w')
outfile.write('['+'\n')
outfile.close()

for name,tuples in d_list.items()[:-1]:
    singlenamedic = {}
    singlenamedic["key"] = name
    singlenamedic["values"] = tuples
    with open('data.json', 'a+') as outfile:
        json.dump(singlenamedic, outfile, sort_keys = True, indent = 4, ensure_ascii=False)
        outfile.write(','+ '\n')

for name,tuples in d_list.items()[-1:]:
    singlenamedic = {}
    singlenamedic["key"] = name
    singlenamedic["values"] = tuples
    with open('data.json', 'a+') as outfile:
        json.dump(singlenamedic, outfile, sort_keys = True, indent = 4, ensure_ascii=False)
        outfile.write('\n')

outfile = open('distance.json', 'a+')
outfile.write('\n'+']')
outfile.close()
        
        
   
    
    
    
    
    
    
    
    
```
